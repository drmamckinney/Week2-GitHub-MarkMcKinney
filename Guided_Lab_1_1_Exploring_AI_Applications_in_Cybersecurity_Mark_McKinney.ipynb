```python
df = pd.read_csv("cybersecurity_login_dataset (1) (2).csv")

df.head()
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nVariables:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

## Step 2 – Dataset Exploration Findings

The dataset contains 600 records and 5 variables:

- login_attempts
- failed_logins
- unusual_login_hour
- new_ip_address
- suspicious

No missing values were identified in the dataset. All variables are stored as integer data types.

```python
print("Suspicious Activity Counts")
print(df["suspicious"].value_counts())

print("\nAverage Failed Logins")
print(df["failed_logins"].mean())

print("\nTop 10 Records with Highest Failed Logins")
print(
    df.sort_values(
        by="failed_logins",
        ascending=False
    ).head(10)
)
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    x="suspicious",
    data=df
)

plt.title("Suspicious vs Normal Activity")
plt.xlabel("Suspicious")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,4))

sns.histplot(
    df["failed_logins"],
    bins=15
)

plt.title("Distribution of Failed Login Attempts")
plt.xlabel("Failed Login Attempts")
plt.ylabel("Frequency")

plt.show()
```

## Suspicious Behavior Analysis

Several records contained unusually high numbers of failed login attempts. These records may indicate brute-force attacks or unauthorized access attempts.

Records containing unusual login hours and new IP addresses may also represent suspicious behavior because they differ from normal user activity.

When multiple indicators occur together, such as high failed logins, unusual login times, and new IP addresses, the likelihood of malicious activity increases and should be investigated by security analysts.

## AI and Cybersecurity

Artificial Intelligence can assist cybersecurity analysts by continuously monitoring login activity and identifying unusual patterns that may indicate malicious behavior.

Patterns that may indicate a security threat include repeated failed login attempts, unusual login times, access from unfamiliar IP addresses, and combinations of multiple risk indicators.

Machine learning can improve cybersecurity monitoring by learning normal user behavior and automatically identifying anomalies that warrant further investigation.

## Discussion Questions

### 1. What login patterns appeared normal?

### 2. What activities appeared suspicious?
